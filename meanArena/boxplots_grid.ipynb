```python
import pandas as pd
from matplotlib import pyplot as plt

dataframe = pd.read_csv("Testing_results/testing_results_grid.csv")
print(dataframe.head())
print(dataframe.isnull().values.any())
```

```python
# Define plots
seven_by_seven = dataframe["7g1m3p3b_o"]
ten_by_ten = dataframe['10g1m3p3b_o']
fifteen_by_fifteen = dataframe['15g1m3p3b_o']
twenty_by_twenty = dataframe['20g1m3p3b_o']

seven_by_seven_ql = dataframe["7g1m3p3b"]
ten_by_ten_ql = dataframe['10g1m3p3b']
fifteen_by_fifteen_ql = dataframe['15g1m3p3b']
twenty_by_twenty_ql = dataframe['20g1m3p3b']

# Apply to arrya so they can be printed to screen
o_columns = [seven_by_seven, ten_by_ten, fifteen_by_fifteen, twenty_by_twenty]
ql_columns = [seven_by_seven_ql, ten_by_ten_ql, fifteen_by_fifteen_ql, twenty_by_twenty_ql]

# arrange the subplots so they can be printed togther
fig, o = plt.subplots()
o.boxplot(o_columns)

fig, ql = plt.subplots()
ql.boxplot(ql_columns)


#titles
o.set_title('Grid - Ajust grid size [7/10/15/20] - Original tallon.py')
o.set_xlabel('Test')
o.set_ylabel('Points Scored')

#titles
ql.set_title('Grid - Ajust grid size [7/10/15/20] - Q-Learning Tallon.py')
ql.set_xlabel('Test')
ql.set_ylabel('Points Scored')

plt.show()
```

```python
import seaborn as sns
sns.boxplot(
    data=[seven_by_seven, seven_by_seven_ql],     
    showmeans=True,
)
plt.title('Seven by Seven Grid')
plt.xlabel("Original Vs Q-Learning")
plt.ylabel("Points Scored")
```

```python
sns.boxplot(
    data=[ten_by_ten, ten_by_ten_ql],
    showmeans=True,
)
plt.title('Ten by Ten Grid')
plt.xlabel("Original Vs Q-Learning")
plt.ylabel("Points Scored")
```

```python
sns.boxplot(
    data=[fifteen_by_fifteen, fifteen_by_fifteen_ql],
    showmeans=True,
)
plt.title('Fifteen by Fifteen Grid')
plt.xlabel("Original Vs Q-Learning")
plt.ylabel("Points Scored")
```

```python
sns.boxplot(
    data=[twenty_by_twenty, twenty_by_twenty_ql],
    showmeans=True,
)
plt.title('Twenty by twenty Grid')
plt.xlabel("Original Vs Q-Learning")
plt.ylabel("Points Scored")
```
