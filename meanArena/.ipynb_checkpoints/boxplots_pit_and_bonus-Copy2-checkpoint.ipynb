```python
import pandas as pd
from matplotlib import pyplot as plt

dataframe = pd.read_csv("Testing_results/testing_results_interval.csv")
print(dataframe.head())
print(dataframe.isnull().values.any())
```

```python
# Define plots
one_interval = dataframe["10g1m1i3p3b_o"]
three_interval = dataframe['10g1m3i3p3b_o']
five_interval = dataframe['10g1m5i3p3b_o']
seven_interval = dataframe['10g1m7i3p3b_o']

one_interval_ql = dataframe["10g1m1i3p3b"]
three_interval_ql = dataframe['10g1m3i3p3b']
five_interval_ql = dataframe['10g1m5i3p3b']
seven_interval_ql = dataframe['10g1m7i3p3b']

# Apply to arrya so they can be printed to screen
o_columns = [one_interval, three_interval, five_interval, seven_interval]
ql_columns = [one_interval_ql, three_interval_ql, five_interval_ql, seven_interval_ql]

# arrange the subplots so they can be printed togther
fig, o = plt.subplots()
o.boxplot(o_columns)

fig, ql = plt.subplots()
ql.boxplot(ql_columns)


#titles
o.set_title('Interval - Ajust Meanie Spawn Interval [1/3/5/7] - Original tallon.py')
o.set_xlabel('Test - 1/3/5/7')
o.set_ylabel('Points Scored')

#titles
ql.set_title('Interval - Ajust Meanie Spawn Interval [1/3/5/7] - Q-Learning tallon.py')
ql.set_xlabel('Test - 1/3/5/7')
ql.set_ylabel('Points Scored')

plt.show()
```

```python
import seaborn as sns
sns.boxplot(
    data=[one_interval, one_interval_ql],     
    showmeans=True,
)
plt.title('One Interval')
plt.xlabel("Original Vs Q-Learning")
plt.ylabel("Points Scored")
```

```python
sns.boxplot(
    data=[three_interval, three_interval_ql],
    showmeans=True,
)
plt.title('Three Interval')
plt.xlabel("Original Vs Q-Learning")
plt.ylabel("Points Scored")
```

```python
sns.boxplot(
    data=[five_interval, five_interval_ql],
    showmeans=True,
)
plt.title('Five Interval')
plt.xlabel("Original Vs Q-Learning")
plt.ylabel("Points Scored")
```

```python
sns.boxplot(
    data=[seven_interval, seven_interval_ql],
    showmeans=True,
)
plt.title('Seven Interval')
plt.xlabel("Original Vs Q-Learning")
plt.ylabel("Points Scored")
```
