```python
import pandas as pd
from matplotlib import pyplot as plt

dataframe = pd.read_csv("Testing_results/testing_results_pit_and_bonus.csv")
print(dataframe.head())
print(dataframe.isnull().values.any())
```

```python
# Define plots
one_bonus_one_pit = dataframe["10g1m1p1b_o"]
five_bonus_five_pit = dataframe['10g1m5p5b_o']
seven_bonus_seven_pit = dataframe['10g1m7p7b_o']


one_bonus_one_pit_ql = dataframe["10g1m1p1b"]
five_bonus_five_pit_ql = dataframe['10g1m5p5b']
seven_bonus_seven_pit_ql = dataframe['10g1m7p7b']

# Apply to arrya so they can be printed to screen
o_columns = [one_bonus_one_pit,five_bonus_five_pit, seven_bonus_seven_pit]
ql_columns = [one_bonus_one_pit_ql, five_bonus_five_pit_ql, seven_bonus_seven_pit_ql]

# arrange the subplots so they can be printed togther
fig, o = plt.subplots()
o.boxplot(o_columns)

fig, ql = plt.subplots()
ql.boxplot(ql_columns)


#titles
o.set_title('Pits and Bonus - Ajust P&M size [1/5/7] - Original tallon.py')
o.set_xlabel('Test - 1/5/7')
o.set_ylabel('Points Scored')

#titles
ql.set_title('Pits and Bonus - Ajust P&M size [1/5/7] - Q-Learning tallon.py')
ql.set_xlabel('Test - 1/5/7')
ql.set_ylabel('Points Scored')

plt.show()
```
